##Загрузите текст.
Текст, который я здесь использую, представляет собой текстовую версию книги Оливера Твиста на английском языке.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag, ne_chunk
from nltk.stem import WordNetLemmatizer
with open('/content/drive/MyDrive/研一/Phc/datatxt/en_OLIVER TWIST.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

##Токенизация.

```python
sentences = sent_tokenize(text)
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]
for i in range(len(sentences)):
    print("Sentence:", sentences[i])
    print("Tokens:", tokenized_sentences[i])
```

##Частеречная разметка.

```python
pos_tagged_sentences = [pos_tag(tokens) for tokens in tokenized_sentences]
for i in range(len(sentences)):
    print("Sentence:", sentences[i])
    print("Tokens:", tokenized_sentences[i])
    print("POS Tags:", pos_tagged_sentences[i])
```

##Лемматизация.


```python
# 词形归并
lemmatizer = WordNetLemmatizer()
lemmatized_sentences = [[lemmatizer.lemmatize(token) for token in tokens] for tokens in tokenized_sentences]
for i in range(len(sentences)):
    # print("Sentence:", sentences[i])
    print("Tokens:", tokenized_sentences[i])
    print("POS Tags:", pos_tagged_sentences[i])
    print("Lemmatized Tokens:", lemmatized_sentences[i])
    # print("Named Entities:", named_entities[i])
    # print()

```

##Выделение (распознавание) именованных сущностей.


```python
# 命名实体识别
named_entities = [ne_chunk(pos_tags) for pos_tags in pos_tagged_sentences]

for i in range(1):
    print("Sentence:", sentences[i])
    print("Tokens:", tokenized_sentences[i])
    print("POS Tags:", pos_tagged_sentences[i])
    print("Lemmatized Tokens:", lemmatized_sentences[i])
    print("Named Entities:", named_entities[i])
    print()
```

##Разбор предложения.

```python
grammar = r"""
    NP: {<DT|JJ|NN.*>+}          # 名词短语
    PP: {<IN><NP>}                # 介词短语
    VP: {<VB.*><NP|PP|CLAUSE>+$}  # 动词短语
    CLAUSE: {<NP><VP>}            # 从句
"""

# 创建语法分析器
chunk_parser = nltk.RegexpParser(grammar)

# 对每个句子进行语法分析
for i in range(len(sentences)):
    print("Sentence:", sentences[i])
    parsed_sentence = chunk_parser.parse(pos_tagged_sentences[i])
    print("Parsed Sentence:", parsed_sentence)
    print()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = "Albert Einstein was a German-born theoretical physicist who developed the theory of relativity."
doc = nlp(text)
print("Tokenization and Part-of-Speech tagging:")
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.ent_type_)
print('-------------------------------------------')
print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
print('------------------------------------------------')
print("\nDependency Parsing:")
for token in doc:
    print(token.text, token.dep_, token.head.text)
```
